```python
import matplotlib.pyplot as plt
import numpy as np

from multisoc.theory.multidimensional_preferences import composite_H
from multisoc.viz import plots
```

```python
fig, ax = plt.subplots(1,5,figsize=(.8*3*6,.8*3))

for i,h in enumerate([.1,.3,.5,.7,.9]):
    h1 = np.array([
        [h,1-h],
        [1-h,h]
    ])
    h2 = h1
    h_mtrx_lst = [h1,h2]
    H_theor = composite_H(
                h_mtrx_lst,
                "all",
                p_d = None,
                alpha = None,
                )
    if i == 4:
        show_clrbr=True
    else:
        show_clrbr=False
    plots.fig_colored_matrix(
	H_theor,
	ax=ax[i],
	xticks=["$mm$","$mM$","$Mm$","$MM$"],
	yticks=["$mm$","$mM$","$Mm$","$MM$"],
	show_colorbar=False,
	figsize=None,
	vmin=0,
	vmax=.81
	)
    plt.title(f"$h={h}$")

plt.savefig(f"temp/H_matrix.png",dpi=600,)

plt.show()
```
